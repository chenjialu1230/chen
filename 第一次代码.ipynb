识别日期

```python
data["Day"] = pd.to_datetime(data["Day"], format="%Y/%m/%d")#大写Y表示四位年份，小写是两位
data
```

数据排序

```python
data = data.sort_values(by = 'Day',ascending = True)#true正序，false倒序
data
```

设置索引index

```python
data = data.set_index('Day',drop=True)#把day扔掉，变成索引
data
```
